### A. Creating Standard Plots (Line, Bar, Pie)

```python
%matplotlib inline
rcParams['figure.figsize'] = (6,4)    #wheather it is in a list or tuple or not makes no difference
sb.set_style('whitegrid')
```

```python
#plot line chart from plot points
x = range(1,10)
y = [2,6,8,3,7,3,9,6,2]

plt.plot(x, y)
```

```python
#plot line chart from pands object
address = 'C:/Users/Arthur/python-learning/mtcars.csv'
data_cars = pd.read_csv(address)
data_cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
mpg = data_cars['mpg']
mpg.plot()
```

```python
#line plot with different variables
diff_vars = data_cars[['cyl','mpg','qsec']]
diff_vars.plot()
```

```python
#plot bar charts from a list input
bar_line = plt.bar(x,y)
#plt.bar(x,y)
```

### C. Plot Formatting

```python
# specify plot color
plt.bar(x,y)
#bar_line
wide = [0.3, 0.3, 0.3, 0.9, 0.9, 0.9, 0.5, 0.5, 0.5]                     #width of bars
color = ['gold']                                                        #color of bars
bar_line = plt.bar(x, y, width=wide, color=color, align='center')       #plot
#bar_line.plot(width=wide, color=color, align='center')
```

```python
#data_color = ['brown', 'red']#, 'yellow', '#A9A9A9', '#FFA07A', '#B0E0E6', '#FFE4C4', '#BDB76B']
data_color = ['brown', 'red', 'yellow', '#A9A9A9', '#FFA07A', '#B0E0E6', '#FFE4C4', '#BDB76B']
diff_vars.plot(color=data_color)
#if the color list is LESS than the number of data items or points, then one or more color would be repeated
#if the color list is MORE than the number of data items or points, then only the first are taken the others are ignored
```

```python
plt.pie(y, colors=data_color)   #note that y has 9 data points y = [2,6,8,3,7,3,9,6,2]; however data_color contains 8 data
plt.show()                      #colors, so that 'brown' is being repeated
```

```python
#Customizing line styles
x1 = range(5,15)
y1 = [9, 4, 8, 7, 6, 5, 9, 3, 2, 5]

plt.plot(x,y)
plt.plot(x1,y1)
```

```python
#Customizing line styles
x1 = range(5,15)
y1 = [9, 4, 8, 7, 6, 5, 9, 3, 2, 5]

plt.plot(x,y, ls='steps', lw=2)
plt.plot(x1,y1, ls='--', lw=4)
```

```python
#Customizing line styles
x1 = range(5,15)
y1 = [9, 4, 8, 7, 6, 5, 9, 3, 2, 5]

plt.plot(x,y, ls='steps', lw=2, marker='*', mew=15)
plt.plot(x1,y1, ls='--', lw=4, marker='1', mew=15)
```

### D. Creating Labels and Annotations

```python
#Labeling plot features
plt.bar(x,y)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
#add labels to a chart using the functional method
z = (100,500,750,900)
phone_sales = ('Nokia','Samsung','Techno','Infinix')
plt.pie(z, labels=phone_sales)
plt.show()

#add legend to a chart using the functional method
plt.pie(z)
plt.legend(phone_sales, loc='upper left')
plt.show()
```

```python
#add labels to a chart using the object oriented method
axes_fig2 = plt.figure()
axes_fig2 = axes_fig2.add_axes([.5,.5,2,2])
axes_fig2.set_xticks(range(32))

mpg.plot()

axes_fig2.set_xticklabels(data_cars.car_names, rotation=60, fontsize='medium')        #rotation not working if mpg.plot 
axes_fig2.set_title('Miles per Gallon for Cars in mtcars')                            #comes after this line
axes_fig2.set_xlabel('car names')                                                     #had to put mpg.plot up
axes_fig2.set_ylabel('miles/gal')

#add legends to a chart using the object oriented method
axes_fig2 = plt.figure()
axes_fig2 = axes_fig2.add_axes([.1,.1,1,1])
axes_fig2.set_xticks(range(32))

mpg.plot()

axes_fig2.set_xticklabels(data_cars.car_names, rotation=80, fontsize='small')        #rotation not working if mpg.plot 
axes_fig2.set_title('Miles per Gallon for Cars in mtcars')                            #comes after this line
axes_fig2.set_xlabel('car names')                                                     #had to put mpg.plot up
axes_fig2.set_ylabel('miles/gal')
axes_fig2.legend(loc='upper left')
```
