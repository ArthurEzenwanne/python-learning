```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import DataFrame, Series

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
```

### A. Creating Standard Plots (Line, Bar, Pie)

```python
%matplotlib inline
rcParams['figure.figsize'] = (6,4)    #wheather it is in a list or tuple or not makes no difference
sb.set_style('whitegrid')
```

```python
#plot line chart from plot points
x = range(1,10)
y = [2,6,8,3,7,3,9,6,2]

plt.plot(x, y)
```

```python
#plot line chart from pands object
address = 'C:/Users/Arthur/python-learning/mtcars.csv'
data_cars = pd.read_csv(address)
data_cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
mpg = data_cars['mpg']
mpg.plot()
```

```python
#line plot with different variables
diff_vars = data_cars[['cyl','mpg','qsec']]
diff_vars.plot()
```

```python
#plot bar charts from a list input
bar_line = plt.bar(x,y)
#plt.bar(x,y)
```

```python
#plot bar charts from a pandas object
mpg.plot(kind='bar')  #for vertical bar charts for a single item mpg
mpg.plot(kind='barh')  #for horizontal bar charts for a single item mpg
diff_vars.plot(kind='bar')  #for vertical bar charts for multiple items mpg, cly, qsec
```

```python
#plot pie charts from a list object
plt.pie(y)
plt.show()         #adding this line only shows the pie chart and removes all other details

plt.pie([2,3,4,5,6])
plt.show()
```

### B. Defining Elements Of A Plot

```python
#defining axes, ticks, and grids
axes_fig = plt.figure().add_axes([.5,.5,1,1])       #not getting the required ticks at the stipulated interval
axes_fig.plot(x,y)
```

```python
axes_fig = plt.figure().add_axes([.5,.5,1,1])       #not getting the required ticks at the stipulated interval

axes_fig.set_xlim([1,9])
axes_fig.set_ylim([0,5])

axes_fig.set_xticks([0,1,2,4,5,6,8,9,10])
axes_fig.set_yticks([2,3,4,5,7,9,11])

axes_fig.plot(x,y)
```

```python
axes_fig = plt.figure().add_axes([.1,.1,1,1])       #not getting the required ticks at the stipulated interval

axes_fig.set_xlim([1,9])
axes_fig.set_ylim([3,7])

#axes_fig.set_xticks([0,1,2,4,5,6,8,9,10])
#axes_fig.set_yticks([2,3,4,5,7,9,11])

#axes_fig.grid()
axes_fig.plot(x,y)
```

```python
# multiple plots in one figure with subplots
axes_fig, (ax1, ax2) = plt.subplots(1,2)    #divivde the axes_fig into 2 parts ax1 and ax2 and place contents with .subplots

ax1.plot(y,x)
ax2.plot(x,y)
```
